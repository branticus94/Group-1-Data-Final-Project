```python
# There are no headers in the data
df = pd.read_excel(r"../4_integrated_csv_files/nhs_data/MHCYP 2023 Data.xlsx", sheet_name='Table 1.2',  header=None)

# Numbert the columns for ease of reference before dividing up and renaming in a more descriptive and helpful way. 
# col1 - col78
new_column_names = [f'col{i+1}' for i in range(78)]

# Assign the new column names to the df
df.columns = new_column_names

# Check this has worked!
print(df.columns)
```

```python
# Split the dataframe into five separate tables based on age groups.
df1 = df[['col1', 'col2', 'col3', 'col4', 'col5', 'col6']]
df2 = df[['col1', 'col7', 'col8', 'col9', 'col10', 'col11']]
df3 = df[['col1', 'col12', 'col13', 'col14', 'col15', 'col16']]
df4 = df[['col1', 'col17', 'col18', 'col19', 'col20', 'col21']]
df5 = df[['col1', 'col22', 'col23', 'col24', 'col25', 'col26']]

# Save these into their own CSV files and renamed with the age groups referenced. 
# table2_11-16 may not be required as we have those ages covered within other tables.
# Note: index=False is used to remove the added index column.
df1.to_csv(r'../6_cleaned_files/nhs_data/table1_8-10.csv', index=False)
df2.to_csv(r'../6_cleaned_files/nhs_data/table2_11-16.csv', index=False)
df3.to_csv(r'../6_cleaned_files/nhs_data/table3_8-16.csv', index=False)
df4.to_csv(r'../6_cleaned_files/nhs_data/table4_17-19.csv', index=False)
df5.to_csv(r'../6_cleaned_files/nhs_data/table3_20-23.csv', index=False)
```
