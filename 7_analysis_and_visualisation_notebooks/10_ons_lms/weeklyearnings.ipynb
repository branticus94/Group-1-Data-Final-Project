```python
import pandas as pd

# Load the dataset
file_path = '../../4_integrated_csv_files/ons_labour_market_statistics/average_weekly_earnings.csv'
data = pd.read_csv(file_path)

# Display the first few rows
print(data.head())
```

```python
data.info()
```

```python
data.describe()
```

```python
data_long = pd.melt(data, id_vars=['sector', 'category', 'unit'],
                    var_name='month', value_name='value')

data_long['month'] = pd.to_datetime(data_long['month'], format='%Y %b')

data_long.head()
```

```python
filtered_data = data_long[(data_long['category'] == 'Total Pay') &
                          (data_long['unit'] == 'pounds £')]

filtered_data.head()
```

```python
filtered_data = filtered_data.copy()

filtered_data['pandemic_period'] = filtered_data['month'].apply(
    lambda x: 'Pre-pandemic' if x < pd.Timestamp('2020-03') else 'Post-pandemic'
)
```

```python
filtered_data = filtered_data.copy()
filtered_data['pandemic_period'] = filtered_data['month'].apply(
    lambda x: 'Pre-pandemic' if x < pd.Timestamp('2020-03') else 'Post-pandemic'
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=filtered_data, x='month', y='value', hue='sector')
plt.title('Average Weekly Earnings Over Time by Sector')
plt.xlabel('Month')
plt.ylabel('Earnings (£)')
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)

plt.savefig('../../8_final_visualisations/ons_employment/weekly_earnings_time_series.png', dpi=300)  

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_data, x='pandemic_period', y='value', hue='sector')
plt.title('Earnings: Pre- vs Post-Pandemic')
plt.xlabel('Pandemic Period')
plt.ylabel('Earnings (£)')
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.savefig('../../8_final_visualisations/ons_employment/earnings_pre_post_pandemic_bar.png', dpi=300)  

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_data, x='pandemic_period', y='value', hue='sector')
plt.title('Weekly Earnings Across Sectors: Pre- vs Post-Pandemic')
plt.xlabel('Pandemic Period')
plt.ylabel('Earnings (£)')
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import matplotlib.dates as mdates

plt.figure(figsize=(12, 6))
sns.lineplot(data=filtered_data, x='month', y='value', hue='pandemic_period', ci=None)
plt.title('Average Weekly Earnings: Pre- vs Post-Pandemic')
plt.xlabel('Year')
plt.ylabel('Earnings (£)')

plt.axvline(pd.Timestamp('2020-03-01'), color='red', linestyle='--', label='Pandemic Start')


plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())  # Show ticks for each year

plt.legend(title='Pandemic Period')
plt.xticks(rotation=45)

plt.savefig('../../8_final_visualisations/ons_employment/weekly_earnings_time_series_2.png', dpi=300)  

plt.show()
```
