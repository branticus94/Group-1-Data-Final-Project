```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_mean_nsa = pd.read_csv(r'../../6_cleaned_files/ons_api_data/measures_of_wellbeing_uk_2017_2023_cleaned.csv')
```

```python
sns.set(style="whitegrid")

# Create the line plot (combined)
plt.figure(figsize=(14, 8))

sns.lineplot(data=df_mean_nsa, x='Time', y='v4_2', hue='Measure_Of_Wellbeing',
             palette=sns.color_palette("viridis"), marker="o")


# Rotate x-axis labels 
plt.xticks(rotation=45)

# Set labels and title
plt.title("Trends in Wellbeing Measures Over Time (Average Mean)", fontsize=16)
plt.xlabel('Time (Year-Quarter)', fontsize=12)
plt.ylabel('Wellbeing Estimate', fontsize=12)

# Show legend
plt.legend(title='Measure of Wellbeing')

plt.savefig(r'../../8_final_visualisations/wellbeing_quarterly/time_series_wellbeing_measures')

# Display the plot
plt.tight_layout()
plt.show()
```
