```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sorted_reset = pd.read_csv(r'../../6_cleaned_files/ons_api_data/suicide_rates_cleaned.csv')
```

```python
# Filtering to look at data for 2017-2023 which is slightly before and after the peak of Covid-19
# and also matches with years available to study in other datasets we have included in our project.

# Filter data to inclide 2017-2023 only
df_filtered = df_sorted_reset[(df_sorted_reset['Year'] >= 2017) & (df_sorted_reset['Year'] <= 2023)]

# Set the plot style
sns.set(style="whitegrid")

# Define a colorblind-safe colour palette 
color_palette = [
    '#440154',  # Purple
    '#20908D',  # Teal
    '#FDE725',  # Yellow
]

# Create the line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(
    x='Year', 
    y='Suicide_Rate', 
    hue='Country', 
    data=df_filtered, 
    marker='o',
    palette=color_palette,
)
    

# Add title and labels
plt.title('Suicide Rate Over Time by Country (2017-2023)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Suicide Rate', fontsize=12)

# Show the plot
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```
