```python
# Remove NaN values
df_cleaned = df.fillna('')
```

```python
"""
Filter referral reasons to EXCLUDE conditions that WOULD NOT be likely to be caused by external factors (although they may have 
  been exacerbated, we cannot speculate): 
  Attachment difficulties
  Bipolar disorder
  Diagnosed Autism
  Organic brain disorder
  Neurodevelopmental Conditions (excluding Autism)
  Personality disorders
  Ongoing or Recurrent Psychosis
   
INCLUDE conditions that are more likely to be influenced by external factors, particularly stress, social isolation, and 
changes in environment (such as during the COVID-19 pandemic or increased social media usage).

 Adjustment to health issues
 Anxiety
 Behaviours that challenge due to a Learning Disability
 Community Perinatal Mental Health Partner Assessment
 Conduct disorders
 Depression
 Drug and alcohol difficulties
 Eating disorders
 Gambling disorder
 Gender Discomfort issues
 In crisis
 Obsessive compulsive disorder (OCD)
 Organic brain disorder (to a lesser extent)
 Perinatal mental health issues

Many of the conditions related to mental health, behavioral difficulties, and substance use are more likely to have been 
exacerbated or influenced by COVID-19 and external social factors (including social media). In contrast, conditions like autism, 
bipolar disorder, and neurodevelopmental conditions have primarily genetic or biological foundations and are not directly caused 
by external circumstances, though their symptoms may be impacted by environmental stressors.
"""

# List of reasons potentially influenced by COVID/social media
external_factors_conditions = [
    'Adjustment to health issues', 'Anxiety', 'Behaviours that challenge due to a Learning Disability', 'Conduct disorders',
    'Depression','Drug and alcohol difficulties','Eating disorders', 'Gambling disorder', 'Gender Discomfort issues','In crisis',
    'Obsessive compulsive disorder (OCD)','Perinatal mental health issues'
]

# Filter the DataFrame to only include relevant conditions
filtered_df = df_cleaned[df_cleaned['Primary Reason for Referral'].isin(external_factors_conditions)]

# Display the filtered DataFrame
print(filtered_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Year': ['2017', '2018', '2019', '2020', '2021'], # adjusted year values for clear plotting
    'Adjustment to health issues': [49293, 81240, 141115, 149455, 199016],
    'Anxiety': [114897, 146988, 206023, 235112, 319718], 
    'Behaviours that challenge due to a Learning Disability': [47, 77, 156, 246, 1903], 
    'Conduct disorders': [27146, 32836, 43364, 42151, 51259],
    'Depression': [164303, 195194, 236958, 225963, 269788],
    'Drug and alcohol difficulties': [19226, 25513, 31078, 32501, 35371],
    'Eating disorders': [26125, 31989, 38164, 50310, 58472],
    'Gambling disorder': [6672, 13356, 15005, 1284, 1454],
    'Gender Discomfort issues': [11365, 8418, 7598, 4401, 2339],
    'In crisis': [327810, 548883, 762195, 975902, 1077377],
    'Obsessive compulsive disorder (OCD)': [5044, 6145, 7262, 7695, 9036],
    'Perinatal mental health issues': [26698, 33618, 47686, 54446, 62541],
}

# Convert to DataFrame
df = pd.DataFrame(data)


# Set the figure size
plt.figure(figsize=(12, 8))

# Plot each condition as a line on the same graph
sns.lineplot(data=df, x='Year', y='Adjustment to health issues', label='Adjustment to Health Issues', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='Anxiety', label='Anxiety', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='Behaviours that challenge due to a Learning Disability', label='Challenging Learning Disabilities', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='Conduct disorders', label='Conduct Disorders', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='Depression', label='Depression', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='Drug and alcohol difficulties', label='Drug & Alcohol Difficulties', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='Eating disorders', label='Eating Disorders', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='Gambling disorder', label='Gambling Disorder', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='Gender Discomfort issues', label='Gender Discomfort', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='In crisis', label='In Crisis', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='Obsessive compulsive disorder (OCD)', label='OCD', marker='o', linewidth=2)
sns.lineplot(data=df, x='Year', y='Perinatal mental health issues', label='Perinatal Mental Health Issues', marker='o', linewidth=2)

# Customise 
from matplotlib.ticker import FuncFormatter

plt.title('Change in Referrals for Conditions Potentially\nAffected by COVID/Social Media (2017-2021)', fontsize=16)
plt.xlabel('Year', fontsize=14)
formatter = FuncFormatter(lambda x, _: f'{int(x):,}')  # change from le6 and add commas as separator
plt.gca().yaxis.set_major_formatter(formatter)
plt.ylabel('Number of Referrals', fontsize=14)
plt.legend(title='Condition', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move the legend to the right

# Display the plot
plt.tight_layout()

plt.savefig(r'../../8_final_visualisations/mental_health/change_in_referrals_for_conditions_2017_2021')

plt.show()

```

```python
# As we can see above, all data increases. Let's look at the percentage increase over the years for
# a more accurate visualisation of the changes


# Data for percentage increases
conditions = [
    'Adjustment to health issues', 'Anxiety', 'Behaviours that challenge due to a Learning Disability',
    'Conduct disorders', 'Depression', 'Drug and alcohol difficulties', 'Eating disorders',
    'Gambling disorder', 'Gender Discomfort issues', 'In crisis', 
    'Obsessive compulsive disorder (OCD)', 'Perinatal mental health issues'
]

percentage_increases = [
    303.6, 178.3, 3942.6, 88.8, 64.2, 83.9, 123.8, -78.1, -79.4, 228.7, 79.2, 134.2
]
custom_palette = {
    'Adjustment to health issues': '#0047cc', # Dark Blue
    'Anxiety': '#ff9933',  # Orange 
    'Behaviours that challenge due to a Learning Disability': '#99cc33',  # Green
    'Conduct disorders': '#ff3366',  # Darker Red  
    'Depression': '#9966ff',  # Purple
    'Drug and alcohol difficulties': '#805500',  # Brown
    'Eating disorders': '#ff80df',  # Pink 
    'Gambling disorder': '#b3b3cc',  # Grey
    'Gender Discomfort issues': '#ffcc00',  # Yellow 
    'In crisis': '#00b3b3',  # Turquoise 
    'Obsessive compulsive disorder (OCD)': '#003399',  # Navy
    'Perinatal mental health issues': '#ff6666',  # Coral 
}

data = pd.DataFrame({'Condition': conditions, 'Percentage Increase': percentage_increases})
data_sorted = data.sort_values(by='Percentage Increase', ascending=False)
# Set the figure size
plt.figure(figsize=(10, 6))

# Create a horizontal bar plot
sns.barplot(data=data_sorted, x='Percentage Increase', y='Condition', palette=custom_palette, hue='Condition')

# Customizing the chart
plt.title('Percentage Increase in Mental \nHealth Referrals (2017-2021)', fontsize=16)
plt.xlabel('Percentage Increase (%)', fontsize=12)
plt.ylabel('Condition', fontsize=12)

# Show the plot
plt.tight_layout()

plt.savefig(r'../../8_final_visualisations/mental_health/horizontal_bar_chart_percentage_increase_in_mental_health_referals')

plt.show()
```
