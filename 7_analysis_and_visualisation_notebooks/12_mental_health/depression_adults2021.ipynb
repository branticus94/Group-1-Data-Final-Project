```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.cm as cm

'''
The depression score is derived using the eight-item Patient Health Questionnaire (PHQ-8) screener. 

Scoring the PHQ-8:
The total score is calculated by adding up the scores for each of the 8 questions. 
The maximum possible score is 24 (if someone rates "nearly every day" on all items), and the minimum possible score is 0 
(if someone rates "not at all" on all items).

Here’s how the score is typically interpreted:

0 to 4: Minimal depression
5 to 9: Mild depression
10 to 14: Moderate depression
15 to 19: Moderately severe depression
20 to 24: Severe depression
'''

```

```python
data_16_24 = df.iloc[2, 0:31] 
data_25_34 = df.iloc[3, 0:31] 
data_35_44 = df.iloc[4, 0:31] 
data_45_54 = df.iloc[5, 0:31] 
data_55_64 = df.iloc[6, 0:31] 
data_65_74 = df.iloc[7, 0:31] 
data_75_over = df.iloc[8, 0:31] 
```

```python
data_16_24
```

```python
# Shorten column renames with only the first date of each range so graph is less messy
new_column_names = {
    '27 to 31 January': '27 January',
    '3 to 7 February': '3 February',
    '10 to 14 February': '10 February',
    '17 to 21 February': '17 February',
    '24 to 28 February': '24 February',
    '3 to 7 March': '3 March',
    '10 to 14 March': '10 March',
    '17 to 21 March': '17 March',
    '24 to 28 March': '24 March',
    '31 March to 4 April': '31 March',
    '7 to 11 April': '7 April',
    '14 to 18 April': '14 April',
    '21 to 25 April': '21 April',
    '28 April to 3 May2': '28 April',
    '5 to 9 May': '5 May',
    '12 to 16 May': '12 May',
    '19 to 23 May': '19 May',
    '26 to 31 May2': '26 May',
    '2 to 6 June': '2 June',
    '9 to 13 June': '9 June',
    '16 to 20 June': '16 June',
    '23 to 27 June': '23 June',
    '30 June to 4 July': '30 June',
    '7 to 11 July': '7 July',
    '14 to 18 July': '14 July',
    '21 to 25 July': '21 July',
    '28 July to 1 August': '28 July',
    '4 to 8 August': '4 August',
    '11 to 15 August': '11 August',
    '18 to 22 August': '18 August'
}

# Renaming the columns
df.rename(columns=new_column_names, inplace=True)

# Check the updated column names
print(df.columns)
```

```python



data_dict = {
    '16 to 24': data_16_24,
    '25 to 34': data_25_34,
    '35 to 44': data_35_44,
    '45 to 54': data_45_54,
    '55 to 64': data_55_64,
    '65 to 74': data_65_74,
    '75 and over': data_75_over
}

# Set up the plot size
plt.figure(figsize=(12, 6))

# Use the viridian color palette for plotting
viridian_palette = sns.color_palette("viridis", n_colors=len(data_dict))

# Define the date ranges (x-axis labels)
date_ranges = df.columns[0:31]  # Columns 0 to 31 represent the dates

# Loop through each age group and plot the data
for idx, (age_group, data) in enumerate(data_dict.items()):
    plt.plot(date_ranges, data, label=age_group, marker='o', color=viridian_palette[idx])

# Add labels and title
plt.xlabel('Weekly Dates')
plt.ylabel('PHQ-8 Score')
plt.title('Weekly Trends in Estimates of Moderate to Severe \nDepressive Symptoms January - August 2021')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Display the legend with age group names
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout for better spacing
plt.tight_layout()

plt.savefig(r'../../8_final_visualisations/mental_health/moderate_to_severe_depressive_symptoms_time_series')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Define the date ranges (x-axis labels)
date_ranges = df.columns[0:31] 

# Set up the plot size
plt.figure(figsize=(12, 6))

# colourblind friendly colour palette
viridian_palette = sns.color_palette("viridis", n_colors=len(data_dict))

# Loop through each age group and plot the data as an area plot
for idx, (age_group, data) in enumerate(data_dict.items()):
    plt.fill_between(date_ranges, data, label=age_group, color=viridian_palette[idx], alpha=0.6)

# Add labels and title
plt.xlabel('Weekly Dates')
plt.ylabel('PHQ-8 Score')
plt.title('Weekly Trends in Estimates of Moderate to Severe \nDepressive Symptoms January - August 2021')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Display the legend with age group names
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout for better spacing
plt.tight_layout()

plt.savefig(r'../../8_final_visualisations/mental_health/moderate_to_severe_depressive_symptoms_time_series_fill_between')

# Show the plot
plt.show()
```
