```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    '2017 (Male)': [17.00, 12.26, 13.28],
    '2020 (Male)': [8.73, 15.31, 20.00],
    '2021 (Male)': [18.73, 15.61, 20.85],
    '2022 (Male)': [10.29, 18.81, 20.29],
    '2023 (Male)': [15.39, 22.32, 17.67],
    '2017 (Female)': [13.43, 14.28, 8.96],
    '2020 (Female)': [27.26, 20.09, 11.98],
    '2021 (Female)': [24.84, 19.84, 14.79],
    '2022 (Female)': [33.08, 21.99, 13.25],
    '2023 (Female)': [31.60, 22.93, 13.55]
}



# Age groups with a logical order (this had to be added because it gave a random age order before)
age_groups = ['17-19', '11-16', '8-10']  

# Create DataFrame with the age_groups as the index
df = pd.DataFrame(data, index=age_groups)

# Create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='viridis', fmt='.2f', linewidths=0.5, cbar_kws={'label': 'Percentage'})

# Customise with label and title
plt.title('Probable Mental Disorders - 8-19 Year Olds in England and Wales 2017-23 \nAge Group, Gender, and Year')
plt.xlabel('Year and Gender')
plt.ylabel('Age Group')

# Show the plot
plt.tight_layout()

plt.savefig('../../8_final_visualisations/nhs/probable_mental_disorders_8_19_heatmap.png', dpi=300)

plt.show()
```
