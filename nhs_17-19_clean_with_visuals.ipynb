```python
import pandas as pd
import os
```

```python
df3 = pd.read_csv("/Users/lisafranchetti/table3_17-19.csv")
```

```python
df3.head(26)
```

```python
df3.rename(columns={'col1': 'age_17-19_years'}, inplace=True)
df3.rename(columns={'col17': '2017_%'}, inplace=True)
df3.rename(columns={'col18': '2020_%'}, inplace=True)
df3.rename(columns={'col19': '2021_%'}, inplace=True)
df3.rename(columns={'col20': '2022_%'}, inplace=True)
df3.rename(columns={'col21': '2023_%'}, inplace=True)
```

```python
# Drop first 5 rows and last rows are they are not required/empty
df3 = df3.drop(index=[0, 1, 2, 3, 4, 26, 27, 28])
df3.head(27)
```

```python
# Convert the columns to numeric, coercing errors to NaN
df3[['2017_%', '2020_%', '2021_%', '2022_%', '2023_%']] = df3[['2017_%', '2020_%', '2021_%', '2022_%', '2023_%']].apply(pd.to_numeric, errors='coerce')
# round down to 2 decimal places
df3[['2017_%', '2020_%', '2021_%', '2022_%', '2023_%']] = df3[['2017_%', '2020_%', '2021_%', '2022_%', '2023_%']].fillna(0).round(2)
# replace 0.00 with blank spaces
df3[['2017_%', '2020_%', '2021_%', '2022_%', '2023_%']] = df3[['2017_%', '2020_%', '2021_%', '2022_%', '2023_%']].replace(0.00, '')
```

```python
# reset the index and the data should now be tidied and easy to understand
df3 = df3.reset_index(drop=True)
df3.head(22)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Year': ['2017', '2020', '2021', '2022', '2023'],
    'Boys/Young Men': [17.0, 8.73, 10.29, 18.73, 15.39], 
    'Girls/Young Women': [13.43, 27.26,	24.84, 33.08, 31.6]  
}

# Create a DataFrame 
df_plot = pd.DataFrame(data)

# Plotting line graph
plt.figure(figsize=(10, 6))
custom_palette = ['#20908D',  # Teal
                  '#FDE725',  # Yellow
                 ]  
sns.lineplot(data=df_plot, x='Year', y='Boys/Young Men', label='Boys', marker='o', color='#20908D')
sns.lineplot(data=df_plot, x='Year', y='Girls/Young Women', label='Girls', marker='o', color='#FDE725')

# Customise
plt.title('Probable Mental Disorders - 17-19 Year Olds in England and Wales 2017-23', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Group')

plt.tight_layout()
plt.show()
```

```python
df_plot = pd.DataFrame(data)

# Reshape the data to long format (for easier plotting)
df_plot_long = df_plot.melt(id_vars='Year', var_name='Group', value_name='Percentage')

# Plotting as a bar chart
plt.figure(figsize=(10, 6))
custom_palette = ['#20908D',  # Teal
                  '#FDE725',  # Yellow
                 ]
sns.barplot(data=df_plot_long, x='Year', y='Percentage', hue='Group', palette=custom_palette, hue_order=['Boys/Young Men', 'Girls/Young Women'])

# Customise
plt.title('Probable Mental Disorders - 17-19 Year Olds in England and Wales 2017-23', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.xticks(rotation=45)
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles=[handles[0], handles[1]], labels=['Boys', 'Girls'], title='Group', loc='upper left')
plt.tight_layout()
plt.show()
```
