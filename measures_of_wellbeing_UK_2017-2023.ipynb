```python
# Remove duplicate columns: 'measure-of-wellbeing', 'wellbeing-estimate', 'seasonal-adjustment'
df = df.drop(['measure-of-wellbeing', 'wellbeing-estimate', 'seasonal-adjustment'], axis=1)
```

```python
df.head()
```

```python
# Split the 'Year' column into 'Year' and 'Quarter'
df[['Year', 'Quarter']] = df['Time'].str.split(' ', expand=True)
print(df)
```

```python
# Remove additional 'yyyy-qq' column as we have the 'Time' column as well as the 'Year' and 'Quarter
df = df.drop('yyyy-qq', axis=1)
```

```python
# Rename columns for clarity and standardisation:
df = df.rename({
    'uk-only': 'UK_Only',
    'MeasureOfWellbeing': 'Measure_Of_Wellbeing',
    'SeasonalAdjustment': 'Seasonal_Adjustment'
    }, axis='columns')
```

```python
df.head()
```

```python
# As we are investigating data slightly before and after the peak of Covid-19, we are
# only interested in results from 2017-2023. We can remove any data relating to years 
# prior to 2017.

years_of_interest = ['2017', '2018', '2019', '2020', '2021', '2022', '2023']

# Filter to keep the rows where the specified areas are in our 'Geography' column 
df_filtered = df[df['Year'].isin(years_of_interest)]

# Display the filtered DataFrame
print(df_filtered)
```

```python
# Reorder columns so 'Year' and 'Quarter' are next to 'Time'
df_filtered=df_filtered[['v4_2', 'LCL', 'UCL', 'Time', 'Year', 'Quarter', 'UK_Only', 'Geography', 'Measure_Of_Wellbeing', 'Estimate', 'Seasonal_Adjustment']]
print(df_filtered)
```

```python
df_filtered.head(25)
```

```python
# Remove NaN and replace with blank spaces
df_filtered = df_filtered.fillna('')
```

```python
# Rearrange data in ascending order from earliest (2017, Q1) to latest (2023, Q4)

# Convert 'Year' to integer
df_filtered['Year'] = df_filtered['Year'].astype(int)

# Convert 'Quarter' to a numerical value (eg: 'Q1'=1, 'Q2'=2)
df_filtered['Quarter'] = df_filtered['Quarter'].astype(str)
df_filtered['Quarter'] = df_filtered['Quarter'].str.extract('(\\d+)').astype(int)
```

```python
df_sorted = df_filtered.sort_values(by=['Year', 'Quarter'], ascending=[True, True])
df_sorted.head(25)
```

```python
# Reset the index
df_sorted_reset = df_sorted.reset_index(drop=True)
```

```python
# Final Review and Summary

# Check for missing values after cleaning
df_sorted_reset.isnull().sum()

# Verify the data types again
df_sorted_reset.info()
```

```python
df_sorted_reset.head(30)
```

```python
# Filter data for 'Average (mean)' in the 'Estimates' column for overall trends over the years in each category:

# 1. Filter for "Average (mean)" entries in the Estimate column.
df_avg_mean = df_sorted_reset[df_sorted_reset['Estimate'] == 'Average (mean)']

# 2. Filter for each of the four specific measures of wellbeing in the Measure_Of_Wellbeing column.
measures_of_wellbeing = ['Anxiety', 'Life satisfaction', 'Happiness', 'Worthwhile']

# 3. Combine both filters to get a dataset with only the 'Average (mean)' values for each measure of wellbeing in order 
#   to visualise our results.

df_mean = df_avg_mean[df_avg_mean['Measure_Of_Wellbeing'].isin(measures_of_wellbeing)]
```

```python
# Display the filtered data
df_mean.head(50)
```

```python
# Filter the Seasonal_Adjustment column to include only 'Non-seasonally adjusted' 
df_mean_nsa = df_mean[df_mean['Seasonal_Adjustment'] == 'Non-seasonally adjusted']
```

```python
sns.set(style="whitegrid")

# Create the line plot (combined)
plt.figure(figsize=(14, 8))

sns.lineplot(data=df_mean_nsa, x='Time', y='v4_2', hue='Measure_Of_Wellbeing',
             palette=sns.color_palette("viridis"), marker="o")


# Rotate x-axis labels 
plt.xticks(rotation=45)

# Set labels and title
plt.title("Trends in Wellbeing Measures Over Time (Average Mean)", fontsize=16)
plt.xlabel('Time (Year-Quarter)', fontsize=12)
plt.ylabel('Wellbeing Estimate', fontsize=12)

# Show legend
plt.legend(title='Measure of Wellbeing')

# Display the plot
plt.tight_layout()
plt.show()
```
